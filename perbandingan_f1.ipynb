```python
import matplotlib.pyplot as plt

data = {
    "Model": ["KNN", "DTree", "Random Forest", "SVM", "XGBoost"],
    "Smote": [0.994141, 0.995150, 0.994175, 0.973555, 0.994163],
    "GAN": [0.992218, 0.992218, 0.991220, 0.970646, 0.992203],
    "WGAN": [0.995131, 0.996101, 0.995131, 0.972603, 0.995131],
}

models = data["Model"]
scenarios = ["Smote", "GAN", "WGAN"]
colors = ["blue", "green", "orange"]

plt.figure(figsize=(10, 6))
for scenario, color in zip(scenarios, colors):
    plt.plot(models, data[scenario], marker="o", label=scenario, color=color)

plt.xlabel("Model")
plt.ylabel("F1-Score")
plt.title("F1-Score Per Model dalam 3 Skenario")
plt.legend(title="Skenario")
plt.ylim(0.97, 1.0)
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```
