```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset.csv', sep = ';')
df_smote=pd.read_csv('smote.csv')
df_gan=pd.read_csv('gan.csv')
df_wgan=pd.read_csv('wgan.csv')
```

```python
fig, axes = plt.subplots(5, 2, figsize=(15, 20))

for i, col in enumerate(df.columns[:-1]):
    row = i // 2 
    col_pos = i % 2 
    
    sns.kdeplot(df[col], label='Real', ax=axes[row, col_pos])
    sns.kdeplot(df_smote[col], label='Smote', ax=axes[row, col_pos])
    sns.kdeplot(df_gan[col], label='GAN', ax=axes[row, col_pos])
    sns.kdeplot(df_wgan[col], label='WGAN', ax=axes[row, col_pos])
    axes[row, col_pos].set_title(f'Real vs Synthetic {col}')
    axes[row, col_pos].legend(title='Target', loc='best')

plt.tight_layout()
plt.show()
```

```python
print("Real Data Statistics:")
display(df.describe().T.style.background_gradient(cmap='coolwarm'))

print("Smote Data Statistics:")
display(df_smote.describe().T.style.background_gradient(cmap='coolwarm'))

print("GAN Data Statistics:")
display(df_gan.describe().T.style.background_gradient(cmap='coolwarm'))

print("WGAN Data Statistics:")
display(df_wgan.describe().T.style.background_gradient(cmap='coolwarm'))
```
